```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import dlib
```

```python
my_image_path = os.getcwd()+'/images/image_03.png'

img_bgr = cv2.imread(my_image_path)    #- OpenCV  
img_bgr = cv2.resize(img_bgr, (640, 500))    # 640x ??  Resize
img_show = img_bgr.copy()      #-    

plt.imshow(img_bgr)
plt.grid(False)
plt.show()
```

#   (face detection)

```python
# dlib  hog detector 
detector_hog = dlib.get_frontal_face_detector()   #- detector 
```

dlib  face detector HOG(Histogram of Oriented Gradient) feature    
 SVM(Support Vector Machine) sliding window  .

```python
# detector   bounding box .
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)
```

dlib detector  dlib.rectangles      
dlib.rectangles  dlib.rectangle    .   
dlib.rectangle  left(), top(), right(), bottom(), height(), width()

```python
print(dlib_rects)   #   

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top() 
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # 

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.grid(False)
plt.show()
```

#  (fade landmark)
-     face landmark localization  . 
- face landmark detection  bounding box  (crop)   .   
   
 
### [ Object keypoint estimation  ]   
Face landmark       object keypoint estimation .       
1) top-down : bounding box  box  keypoint    
2) bottom-up :   keypoint   point     box 

 68  ibug . Dlib ibug 300-W   pretrained model  .

```python
#  landmark  
model_path = os.getcwd()+'/model/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
```

```python
# landmark_predictor RGB  dlib.rectangle   dlib.full_object_detection  .
# points dlib.full_object_detection    parts()      .

list_landmarks = []

for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))
```

```python
#    .

for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1) # yellow

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.grid(False)
plt.show()
```

#  
    2 .

2.   : weight = height

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[30]) # left eye bottom index : 40
    x = landmark[40][0] 
    y = landmark[40][1]
    w = dlib_rect.width()
    h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
```

```python
#      .

sticker_path = os.getcwd()+'/images/cat-whiskers.png'

img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)

img_sticker_rgb = cv2.cvtColor(img_sticker, cv2.COLOR_BGR2RGB)
plt.imshow(img_sticker_rgb)
plt.show
```

```python
#       x, y  . 
#   top-left .

refined_x = x - w // 3 # left
refined_y = y - h // 3 # top
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:]
    refined_y = 0
```

```python
#    
# sticker_area     crop  
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.grid(False)
plt.show()
```

```python
# bounding box landmark     . 
# img_show  img_rgb .

sticker_area = img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.grid(False)
plt.show()
```

```python
#   


sticker_area = img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    cv2.addWeighted(sticker_area, 0.3, np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8), 0.7, 0)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.grid(False)
plt.show()
```
